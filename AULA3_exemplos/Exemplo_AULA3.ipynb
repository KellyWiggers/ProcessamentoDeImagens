# Operações pontuais

Realizar o import da biblioteca Opencv. 

```python
import cv2 #importa a biblioteca
```

As imagens no OpenCV, por padrão, são trabalhadas no formato BGR (blue, green, red), ao invés de RGB. Portanto, tome cuidado quando for manipular, modificar cores, para não se confundir da ordem da biblioteca. 

```python
img = cv2.imread( 'lena.png') #abrir a imagem
```

```python
print(img) #Mostra os pixels
```

Se tiver usando Python + Opencv apenas faça:

cv2.imshow("Nome da janela", img) 

para mostrar a imagem a com a função imshow em uma nova janela

No jupyter notebook podemos usar o matplotlib para plotar a imagem dentro do próprio notebook.

```python
from matplotlib import pyplot as plt
```

Visualizando a imagem lida

```python
plt.imshow(img)
```

Contudo, como o matplotlib trabalha diretamente com RGB, e o padrão do OpenCV é BGR, precisamos fazer uma conversão. 

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

Agora sim, temos a imagem sendo plotada com as cores originais. 

Podemos verificar as dimensões da imagem. Quantos pixels na largura e na altura. bem como o número de canais.

```python
#Dimensoes
height = img.shape[0] #altura
width = img.shape[1] #largura  
channels = img.shape[2] #numero de canais
```

```python
print('Largura em pixels: {}'.format(width))
print('Altura em pixels: {}'.format(height))
print('Qtde de canais: {}'.format(channels))
```

Podemos também acessar pixels específicos e verificar as cores daquele pixel. Por exemplo, vamos acessar o pixel da posição linha 0 e coluna 0.

```python
px = img[0, 0] #acessando valor de pixel posicao 0 e 0 (h,w)
print('O pixel (0, 0) tem as seguintes cores:')
print(px)
```

```python
(b, g, r) = img[0,0] #acessando pixel separado
print('Vermelho:', r, 'Verde:', g, 'Azul:', b)
```

Além disso, modificações podem ser realizadas apenas realizando operações com matrizes. Veja como podemos modificar a posição linha 99 e coluna 51, e transformar esse pixel para azul.

```python
#Acessando posição h e w e muda pra azul. 
img[51, 99] =  (255,0,0)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

Note que no meio do chapeu tem um pixel todo azul. 

Vamos abrir a imagem novamente, e agora criar um retângulo vermelho sobre ela. Ao invés de fazer um for, podemos simplesmente acessar os limites de largura e comprimento, adicionando a nova cor do pixel. Veja o exemplo abaixo:

```python
img = cv2.imread( 'lena.png') #abrir a imagem
img[100:150, 50:100] = (0, 0, 255)
```

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

Para salvar uma imagem no disco, use o seguinte comando:

```python
#Salvar a imagem no disco com função imwrite()
cv2.imwrite("saida.jpg", img)
```

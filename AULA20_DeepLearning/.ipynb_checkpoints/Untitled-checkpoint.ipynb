### Tensor

Um tensor  como se fosse uma matriz, a diferena  que suas dimenses no esto limitadas a 2. Ele pode tanto ter dimenso 0 (escalar) quanto dimenso n.





Precisamos converter os dados para tensor, pois  assim que a biblioteca trabalha com as informaes.

```python
import numpy as np
import torch
import torch.nn.functional as F
import torchvision
import matplotlib.pyplot as plt
from time import time
from torchvision import datasets, transforms
from torch import nn, optim

transform = transforms.ToTensor() #definindo a converso de imagem para tensor

trainset = datasets.MNIST('./MNIST_data/', download=True, train=True, transform=transform) # Carrega a parte de treino do dataset
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True) # Cria um buffer para pegar os dados por partes

valset = datasets.MNIST('./MNIST_data/', download=True, train=False, transform=transform) # Carrega a parte de validao do dataset
valloader = torch.utils.data.DataLoader(valset, batch_size=64, shuffle=True) # Cria um buffer para pegar os dados por partes
```

* a partir da transforms.ToTensor() definimos a varivel transform como a converso dos dados do dataset para tensor.
* carregamos o set de treino na varivel trainset. o primeiro argumento  o local onde ser salvado o dataset, train=true determina que voc quer carregar a parte reservada para treino (85,7% dos dados) e no ltimo argumento, passamos a transformao definida anteriormente.
* por ser um dataset muito grande(70.000 imagens), devemos criar um buffer para otimizar a tarefa e acess-lo em subgrupos menores. Para isso, definimos a varivel trainloader que  um objeto itervel que nos fornece lotes(batchs) de itens at que o set seja todo percorrido. Passamos como argumento o trainset, batch_size=64 para que os subgrupos sejam de 64 itens e shuffle=true que determina que iremos tomar os itens de forma aleatria, ou seja, um mesmo item pode estar presente em mais de um subgrupo.
* repetimos o mesmo processo para o set de validao

### Importar a base
Esse tutorial usa a base de dados Fashion MNIST que contm 70,000 imagens em tons de cinza em 10 categorias. As imagens mostram artigos individuais de roupas com baixa resoluo (28 por 28 pixels), como vemos aqui:



```python
fashion_mnist = keras.datasets.fashion_mnist

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

Fashion MNIST tem como inteno substituir a clssica base de dados MNIST frequentemente usada como "Hello, World" de programas de aprendizado de mquina (machine learning) para viso computacional. A base de dados MNIST contm imagens de dgitos escritos  mo (0, 1, 2, etc.) em um formato idntico ao dos artigos de roupas que usaremos aqui.

Esse tutorial usa a Fashion MNIST para variar, e porque  um problema um pouco mais desafiador que o regular MNIST. Ambas bases so relativamente pequenas e so usadas para verificar se um algoritmo funciona como esperado. Elas so bons pontos de partida para testar e debugar cdigo.

Usaremos 60,000 imagens para treinar nossa rede e 10,000 imagens para avaliar quo precisamente nossa rede aprendeu a classificar as imagens. Voc pode acessar a Fashion MNIST diretamente do TensorFlow. Importe e carregue a base Fashion MNIST diretamente do TensorFlow:

Carregando a base de dados que retorna quatro NumPy arrays:

* Os arrays train_images e train_labels so o conjunto de treinamento os dados do modelo usados para aprender.
* O modelo  testado com o conjunto de teste, os arrays test_images e test_labels.
As imagens so arrays NumPy de 28x28, com os valores de pixels entre 0 to 255. As labels (alvo da classificao) so um array de inteiros, no intervalo de 0 a 9. Esse corresponde com a classe de roupa que cada imagem representa:

Cada imagem  mapeada com um s label. J que o nome das classes no so includas na base de dados, armazene os dados aqui para us-los mais tarde quando plotarmos as imagens:

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

### Pr-processe os dados

Os dados precisam ser pr-processados antes de treinar a rede. Se voc inspecionar a primeira imagem do conjunto de treinamento, voc ver que os valores dos pixels esto entre 0 e 255:

```python
plt.figure()
plt.imshow(train_images[0])
plt.colorbar()
plt.grid(False)
plt.show()
```

Escalaremos esses valores no intervalo de 0 e 1 antes de alimentar o modelo da rede neural. Para fazer isso, dividimos os valores por 255.  importante que o conjunto de treinamento e o conjunto de teste podem ser pr-processados do mesmo modo:

```python
train_images = train_images / 255.0

test_images = test_images / 255.0
```

Para verificar que os dados esto no formato correto e que estamos prontos para construir e treinar a rede, vamos mostrar as primeiras 25 imagens do conjunto de treinamento e mostrar o nome das classes de cada imagem abaixo.

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
plt.show()
```

### Construindo o modelo

Construir a rede neural requer configurar as camadas do modelo, e depois, compilar o modelo.

#### Montar as camadas
O principal bloco de construo da rede neural  a camada (layer). As camadas (layers) extraem representaes dos dados inseridos na rede. Com sorte, essas representaes so significativas para o problema  mo.

Muito do deep learning consiste em encadear simples camadas. Muitas camadas, como tf.keras.layers.Dense, tem parmetros que so aprendidos durante o treinamento.

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
```

A primeira camada da rede, tf.keras.layers.Flatten, transforma o formato da imagem de um array de imagens de duas dimenses (of 28 by 28 pixels) para um array de uma dimenso (de 28 * 28 = 784 pixels). Pense nessa camada como camadas no empilhadas de pixels de uma imagem e os enfilere. Essa camada no tem parmetros para aprender; ela s reformata os dados.

Depois dos pixels serem achatados, a rede consiste de uma sequncia de duas camadas tf.keras.layers.Dense. Essas so camadas neurais densely connected, ou fully connected. A primeira camada Dense tem 128 ns (ou neurnios). A segunda (e ltima) camada  uma softmax de 10 ns que retorna um array de 10 probabilidades, cuja soma resulta em 1. Cada n contm um valor que indica a probabilidade de que aquela imagem pertence a uma das 10 classes.

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

### Compile o modelo
Antes do modelo estar pronto para o treinamento,  necessrio algumas configuraes a mais. Essas sero adicionadas no passo de compilao:

* Funo Loss  Essa mede quo precisa o modelo  durante o treinamento. Queremos minimizar a funo para guiar o modelo para a direo certa.
* Optimizer  Isso  como o modelo se atualiza com base no dado que ele v e sua funo loss.
* Mtricas  usadas para monitorar os passos de treinamento e teste. O exemplo abaixo usa a acurcia, a frao das imagens que foram classificadas corretamente.

### Treine o modelo
Treinar a rede neural requer os seguintes passos:

1. Alimente com os dados de treinamento, o modelo. Neste exemplo, os dados de treinamento so os arrays train_images e train_labels.
2. O modelo aprende como associar as imagens as labels.
3. Perguntamos ao modelo para fazer previses sobre o conjunto de teste  nesse exemplo, o array test_images. Verificamos se as previses combinaram com as labels do array test_labels.

Para comear a treinar, chame o mtodo model.fit assim chamado, porque ele "encaixa" o modelo no conjunto de treinamento:

```python
model.fit(train_images, train_labels, epochs=10)
```

 medida que o modelo treina, as mtricas loss e acurcia so mostradas. O modelo atinge uma acurcia de 0.88 (ou 88%) com o conjunto de treinamento.

### Avalie a acurcia
Depois, compare como o modelo performou com o conjunto de teste:

```python
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)

print('\nTest accuracy:', test_acc)
```

Acabou que o a acurcia com o conjunto de teste  um pouco menor do que a acurcia de treinamento. Essa diferena entre as duas acurcias representa um overfitting. Overfitting  modelo de aprendizado de mquina performou de maneira pior em um conjunto de entradas novas, e no usadas anteriormente, que usando o conjunto de treinamento.

### Faa predies
Com o modelo treinado, o usaremos para predies de algumas imagens.

```python
predictions = model.predict(test_images)
```

Aqui, o modelo previu que a label de cada imagem no conjunto de treinamento. Vamos olhar na primeira predio:

```python
predictions[0]
```

A predio  um array de 10 nmeros. Eles representam um a confiana do modelo que a imagem corresponde a cada um dos diferentes artigos de roupa. Podemos ver cada label tem um maior valor de confiana:

```python
np.argmax(predictions[0])
```

Ento, o modelo  confiante de que esse imagem  uma bota (ankle boot) ou class_names[9]. Examinando a label do teste, vemos que essa classificao  correta:



```python
test_labels[0]
```

Podemos mostrar graficamente como se parece em um conjunto total de previso de 10 classes.

```python
def plot_image(i, predictions_array, true_label, img):
    predictions_array, true_label, img = predictions_array[i], true_label[i], img[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])

    plt.imshow(img, cmap=plt.cm.binary)

    predicted_label = np.argmax(predictions_array)
    if predicted_label == true_label:
        color = 'blue'
    else:
        color = 'red'

    plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                100*np.max(predictions_array),
                                class_names[true_label]),
                                color=color)

def plot_value_array(i, predictions_array, true_label):
    predictions_array, true_label = predictions_array[i], true_label[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    thisplot = plt.bar(range(10), predictions_array, color="#777777")
    plt.ylim([0, 1])
    predicted_label = np.argmax(predictions_array)

    thisplot[predicted_label].set_color('red')
    thisplot[true_label].set_color('blue')
```

Vamos olhar a previso imagem na posio 0, do array de predio.

```python
i = 0
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plot_image(i, predictions, test_labels, test_images)
plt.subplot(1,2,2)
plot_value_array(i, predictions,  test_labels)
plt.show()
```

```python
i = 12
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plot_image(i, predictions, test_labels, test_images)
plt.subplot(1,2,2)
plot_value_array(i, predictions,  test_labels)
plt.show()
```

Vamos plotar algumas da previso do modelo. Labels preditas corretamente so azuis e as predies erradas so vermelhas. O nmero d a porcentagem (de 100) das labels preditas. Note que o modelo pode errar mesmo estando confiante.

```python
# Plota o primeiro X test images, e as labels preditas, e as labels verdadeiras.
# Colore as predies corretas de azul e as incorretas de vermelho.
num_rows = 5
num_cols = 3
num_images = num_rows*num_cols
plt.figure(figsize=(2*2*num_cols, 2*num_rows))
for i in range(num_images):
    plt.subplot(num_rows, 2*num_cols, 2*i+1)
    plot_image(i, predictions, test_labels, test_images)
    plt.subplot(num_rows, 2*num_cols, 2*i+2)
    plot_value_array(i, predictions, test_labels)
plt.show()
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

No próximo exemplo, incluímos as variáveis de linha e coluna para serem as componentes de cor, lembrando que as variáveis componentes da cor devem assumir o valor entre 0 e 255 então utilizamos a operação “resto da divisão por 256” para manter o resultado entre 0 e 255.

```python
for y in range(0, img.shape[0]): #percorre linhas
     for x in range(0, img.shape[1]): #percorre colunas
             img[y, x] = (x%256,y%256,x%256)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

Agora no próximo exemplo utilizamos os valores de linha multiplicado pela coluna (x*y) no componente “G” da tupla que forma a cor de cada pixel e deixamos o componente azul e
vermelho zerados

```python
for y in range(0, img.shape[0], 1): #percorre as linhas
     for x in range(0, img.shape[1], 1): #percorre as colunas
             img[y, x] = (0,(x*y)%256,0)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

O objetivo agora é saltar a cada 10 pixels ao percorrer as linhas e mais 10 pixels ao percorrer as colunas. A cada salto é
criado um quadrado amarelo de 5x5 pixels

```python
for y in range(0, img.shape[0], 10): #percorre linhas
    for x in range(0, img.shape[1], 10): #percorre colunas
         img[y:y+5, x: x+5] = (0,255,255)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

## Redimensionamento de uma imagem

```python
img = cv2.imread( 'lena.png') 
largura = img.shape[1]
altura = img.shape[0]
proporcao = float(altura/largura)
largura_nova = 520 #em pixels
altura_nova = int(largura_nova*proporcao)
tamanho_novo = (largura_nova, altura_nova)
img_redimensionada = cv2.resize(img, tamanho_novo, interpolation = cv2.INTER_AREA) 
plt.imshow(cv2.cvtColor(img_redimensionada, cv2.COLOR_BGR2RGB))
```

## Redimensionamento utilizando slicing. 

O código basicamente refaz a imagem interpolando linhas e colunas, ou seja, pega a primeira linha, ignora a segunda, depois pega a terceira linha, ignora a quarta, e assim por
diante. O mesmo é feito com as colunas. Então temos uma imagem que é exatamente ¼ (um quarto) da original, tendo a metade da altura e a metade da largura da orginal.

```python
img = cv2.imread( 'lena.png') 
img_redimensionada = img[::2,::2]
plt.imshow(cv2.cvtColor(img_redimensionada, cv2.COLOR_BGR2RGB))
```

Note a qualidade da imagem redimensionada.

## Rotacionando uma imagem

Para isso, precisamos encontrar o centro da imagem, e então rotacionar quandos graus forem necessários, sentido horário ou anti-horário. 
 
 A transformação affine ou mapa affine, é uma função entre espaços affine que preservam os pontos, grossura de linhas e planos. Além disso, linhas paralelas permanecem paralelas após uma transformação affine. Essa transformação não necessariamente preserva a distância entre pontos mas ela preserva a proporção das distâncias entre os pontos de uma linha reta. Uma rotação é um tipo de transformação affine

```python
img = cv2.imread( 'lena.png') 
(alt, lar) = img.shape[:2]
centro = (lar // 2, alt // 2) 
M = cv2.getRotationMatrix2D(centro, 30, 1.0) #30 graus
img_rotacionada = cv2.warpAffine(img, M, (lar, alt))
plt.imshow(cv2.cvtColor(img_rotacionada, cv2.COLOR_BGR2RGB))
```

```python
img = cv2.imread( 'lena.png') 
(alt, lar) = img.shape[:2]
centro = (lar // 2, alt // 2) 
M = cv2.getRotationMatrix2D(centro, 90, 1.0) #30 graus
img_rotacionada = cv2.warpAffine(img, M, (lar, alt))
plt.imshow(cv2.cvtColor(img_rotacionada, cv2.COLOR_BGR2RGB))
```
